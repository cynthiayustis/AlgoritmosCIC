```python
import random
import numpy as np
import copy
import math
import pygame,sys
from pygame.locals import *

class Nodo:
    def __init__(self, nombre):
        self.nombre = nombre
        self.conexiones = list()
        self.grado = 0
        self.revisado = False
        self.nivel = 0
        self.ancestro = None
        self.peso = float('inf')
        self.coordenadas = [0,0]
    def addNode(self,nuevo_nodo):
        if not(nuevo_nodo in self.getConections()):
            self.conexiones.append(nuevo_nodo)
    def getName(self):
        return self.nombre
    def addAncestro(self,nodo_ancestro):
        self.ancestro = nodo_ancestro
    def getAncestro(self):
        return self.ancestro
    def addCoordenadas(self,x,y):
        self.coordenadas = [x,y]
    def getCoordenadas(self):
        return self.coordenadas
    def addRevisado(self,revisado):
        self.revisado = revisado
    def addNivel(self,level):
        self.nivel = level
    def getNivel(self):
        return self.nivel
    def getRevisado(self):
        return self.revisado
    def getConections (self): #funcion para obtener la lista de conexiones
        return self.conexiones
    def addGrado(self):
        self.grado = self.grado + 1
    def getGrado(self):
        return self.grado
    def addPeso(self,peso):
        self.peso = peso
    def getPeso(self):
        return self.peso
```

```python
class Grafo:
    def __init__(self, nombre):
        self.nombre = nombre

    # Función para verificar que las aristas no se repitan
    def aristaRepetida(lista, ancestro, descendiente):
        repetido = 0
        for arista in lista:
            nodo1 = arista.getAncestro()
            nodo2 = arista.getDescendiente()
            if nodo1 == ancestro: 
                  if nodo2 == descendiente:
                    repetido = 1
            if nodo2 == ancestro:
                  if nodo1 == descendiente:
                    repetido = 1
        return repetido


    #Función para guardar matriz
    def GuardarMatriz(nombre,aristas,n):
            file = open(nombre + "_" +str(n)+ ".gv", "w")
            #file = open(nombre,"w") #a para anexar al anterior texto, w para sobreescribir
            file.write("graph abstract {\n")
            for arista in aristas:
                nodo1 = arista.getAncestro().getName()
                nodo2 = arista.getDescendiente().getName()
                file.write("nodo_" + nodo1 + "->nodo_" + nodo2 + ";\n")
            file.write("}")

    #Función para guardar matriz
    def GuardarMatriz2(nombre,arbol):
            file = open(nombre + ".gv", "w")
            #file = open(nombre,"w") #a para anexar al anterior texto, w para sobreescribir
            file.write("graph abstract {\n")
            for arista in arbol:
                nodo1 = arista.getAncestro().getName()
                #print(nodo1)
                nodo2 = arista.getDescendiente().getName()
                #print(nodo2)
                file.write("nodo_" + nodo1 + "->nodo_" + nodo2 + ";\n")
            file.write("}")
            


    def GuardarMatriz2_Dijkstra(nombre,arbol):
            file = open(nombre + ".gv", "w")
            #file = open(nombre,"w") #a para anexar al anterior texto, w para sobreescribir
            file.write("graph abstract {\n")
            for arista in arbol:
                nodo1 = arista.getAncestro().getName()
                #print(nodo1)
                nodo2 = arista.getDescendiente().getName()
                #print(nodo2)
                peso = arista.getDescendiente().getPeso()
                peso2 = arista.getAncestro().getPeso()
                file.write("nodo_" + nodo1 + '('+ str(peso2) + ')' + "->nodo_" + nodo2 +'('+ str(peso) + ')' + ";\n")
            file.write("}")
            


    #Función para recontruir el arbol despues de las busquedas
    def reconstruir_arbol(nodos):
            aristas = list()
            k = 1
            for nodo in nodos:
                ancestro = nodo.getAncestro()
                if ancestro != None:
                    #print(ancestro.getName())
                    aristas.append(Arista(str(k)))
                    arista = aristas[k-1]
                    arista.addAncestro(ancestro)
                    arista.addDescendiente(nodo)
                    k += 1
            return aristas



    # Modelo G(m,n) de malla
    def Malla(n,m):
            nodos = list() #Creamos una lista vacía donde se guardarán todos los nodos
            aristas = list() #Creamos una lista vacía donde se guardarán todas las aristas
            #aristas.append(0)
            k = 1
            for i in range(1, m*n+1): #creamos los n nodos
                nodos.append(Nodo(str(i)))
            k = 1
            for i in range(n-1):
                for j in range(m-1):
                    pos = i*m+j
                    nodo = nodos[pos]
                    nodo1 = nodos[pos]
                    nodo2 = nodos[j+(i+1)*m]
                    nodo3 = nodos[pos+1]
                    aristas.append(Arista(str(k)))
                    arista = aristas[k-1]
                    arista.addAncestro(nodo1)
                    arista.addDescendiente(nodo2)
                    nodo1.addNode(nodo2)
                    nodo2.addNode(nodo1)
                    arista.addPeso(random.randint(1, 100))
                    k += 1  
                    aristas.append(Arista(str(k)))
                    arista = aristas[k-1]
                    arista.addAncestro(nodo1)
                    arista.addDescendiente(nodo3)
                    nodo3.addNode(nodo1)
                    nodo1.addNode(nodo3)
                    arista.addPeso(random.randint(1, 100))
                    k += 1
            Grafo.GuardarMatriz("Malla",aristas,n)
            return aristas,nodos



    # Modelo Gn,m de Erdös y Rényi  
    def ErdosRenyi(n,m): # n nodos, m aristas
            nodos = list() #Creamos una lista vacía donde se guardarán todos los nodos
            aristas = list() #Creamos una lista vacía donde se guardarán todas las aristas
            for i in range(1, n+1): #creamos los n nodos
                nodos.append(Nodo(str(i)))
            #Inicia el ciclo de Erdos Renyi
            j = 1
            while j <= m and j <= n*(n-1)/2: # j menor que el número de aristas
                num1 = random.randint(0, n-1)
                num2 = random.randint(0, n-1)
                repetido = Grafo.aristaRepetida(aristas, nodos[num1],nodos[num2])
                if num1 != num2 and repetido == 0 :
                    aristas.append(Arista(str(j)))
                    nodo1 = nodos[num1]
                    nodo2 = nodos[num2]
                    nueva_arista = aristas[j-1]
                    nueva_arista.addPeso(random.randint(1, 100))
                    nueva_arista.addAncestro(nodo1)
                    nueva_arista.addDescendiente(nodo2)
                    nodo1.addNode(nodo2)
                    nodo2.addNode(nodo1)
                    j += 1
            Grafo.GuardarMatriz("ErdosRenyi",aristas,n)
            return aristas,nodos



    # Modelo Gn,p de Gilbert
    def Gilbert(n,p):
            nodos = list() #Creamos una lista vacía donde se guardarán todos los nodos
            aristas = list() #Creamos una lista vacía donde se guardarán todas las aristas
            for i in range(1, n+1): #creamos los n nodos
                nodos.append(Nodo(str(i)))
            k = 1
            for i in range(n):
                for j in range(n):
                    prob = random.random()
                    if prob < p and i != j:
                        nodo1 = nodos[i]
                        nodo2 = nodos[j]
                        repetido = Grafo.aristaRepetida(aristas, nodos[i],nodos[j])
                        if repetido == 0:
                            aristas.append(Arista(str(k)))
                            nueva_arista = aristas[k-1]
                            nueva_arista.addPeso(random.randint(1, 100))
                            nueva_arista.addAncestro(nodo1)
                            nueva_arista.addDescendiente(nodo2)
                            nodo1.addNode(nodo2)
                            nodo2.addNode(nodo1)
                            k += 1
            Grafo.GuardarMatriz("Gilbert",aristas,n)
            return aristas,nodos



    # Modelo Gn,r geográfico simple
    def Geografico(n,r):
            nodos = list() #Creamos una lista vacía donde se guardarán todos los nodos
            aristas = list() #Creamos una lista vacía donde se guardarán todas las aristas
            for i in range(1, n+1): #creamos los n nodos
                nodos.append(Nodo(str(i)))
            #Ahora debemos generar las coordenadas para cada nodo en el cuadro unitario
            x = np.random.normal(0, 0.5, n)
            y = np.random.normal(0, 0.5, n)
            k = 1
            for i in range(n):
                for j in range(i,n):
                    x1 = x[i]
                    x2 = x[j]
                    y1 = y[i]
                    y2 = y[j]
                    distancia = np.sqrt((x1-x2)**2 + (y1-y2)**2)
                    if distancia < r and i != j:
                        nodo1 = nodos[i]
                        nodo2 = nodos[j]
                        repetido = Grafo.aristaRepetida(aristas, nodos[i],nodos[j])
                        if repetido == 0:
                            aristas.append(Arista(str(k)))
                            nueva_arista = aristas[k-1]
                            nueva_arista.addPeso(random.randint(1, 100))
                            nueva_arista.addAncestro(nodo1)
                            nueva_arista.addDescendiente(nodo2)
                            nodo1.addNode(nodo2)
                            nodo2.addNode(nodo1)
                            k += 1
            Grafo.GuardarMatriz("Geografico",aristas,n)
            return aristas,nodos



    # Modelo Gn,d Barabási-Albert.
    def BarabasiAlbert(n,d):
            nodos = list() #Creamos una lista vacía donde se guardarán todos los nodos
            aristas = list() #Creamos una lista vacía donde se guardarán todas las aristas
            k = 1
            for i in range(1, n+1): #creamos los n nodos
                nodos.append(Nodo(str(i)))
            for i in range(n):
                    nodo1 = nodos[i]
                    for j in range(i):
                        nodo2 = nodos[j]
                        grado = nodo2.getGrado()
                        p = 1 - grado/d
                        #print("Nodo " + nodo2.getName() + " grado " + str(grado))
                        prob = random.random()
                        if p > prob:
                            aristas.append(Arista(str(k)))
                            arista = aristas[k-1]
                            arista.addPeso(random.randint(1, 100))
                            arista.addAncestro(nodo2)
                            arista.addDescendiente(nodo1)
                            nodo1.addNode(nodo2)
                            nodo2.addNode(nodo1)
                            nodo1.addGrado()
                            nodo2.addGrado()
                            k += 1
            Grafo.GuardarMatriz("BarabasiAlbert",aristas,n)
            return aristas,nodos


    # Modelo Gn Dorogovtsev-Mendes
    def Dorogovtsev_Mendes(n):
            nodos = list() #Creamos una lista vacía donde se guardarán todos los nodos
            aristas = list() #Creamos una lista vacía donde se guardarán todas las aristas
            for i in range(1, n+1): #creamos los n nodos
                nodos.append(Nodo(str(i)))
            nodo1 = nodos[0]
            nodo2 = nodos[1]
            nodo3 = nodos[2]
            aristas.append(Arista(str(1)))
            arista = aristas[0]
            arista.addPeso(random.randint(1, 100))
            arista.addAncestro(nodo1)
            arista.addDescendiente(nodo2)
            nodo1.addNode(nodo2)
            nodo2.addNode(nodo1)
            aristas.append(Arista(str(2)))
            arista = aristas[1]
            arista.addPeso(random.randint(1, 100))
            arista.addAncestro(nodo2)
            arista.addDescendiente(nodo3)
            nodo3.addNode(nodo2)
            nodo2.addNode(nodo3)
            aristas.append(Arista(str(3)))
            arista = aristas[2]
            arista.addPeso(random.randint(1, 100))
            arista.addAncestro(nodo3)
            arista.addDescendiente(nodo1)
            nodo3.addNode(nodo1)
            nodo1.addNode(nodo3)
            if n > 3:
                k = 4
                for i in range(3,n):
                    num = random.randint(0,len(aristas)-1) #generamos un número aleatorio de arista
                    arista = aristas[num]  #seleccionamos esa arista
                    nodo1 = arista.getAncestro()
                    nodo2 = arista.getDescendiente()
                    nodo3 = nodos[i]
                    aristas.append(Arista(str(k)))
                    arista = aristas[k-1]
                    arista.addPeso(random.randint(1, 100))
                    arista.addAncestro(nodo1)
                    arista.addDescendiente(nodo3)
                    nodo3.addNode(nodo1)
                    nodo1.addNode(nodo3)
                    k += 1
                    aristas.append(Arista(str(k)))
                    arista = aristas[k-1]
                    arista.addPeso(random.randint(1, 100))
                    arista.addAncestro(nodo2)
                    arista.addDescendiente(nodo3)
                    nodo2.addNode(nodo3)
                    nodo3.addNode(nodo2)
                    k += 1
            Grafo.GuardarMatriz("Dorogovtsev-Mendes",aristas,n)
            return aristas,nodos


    # Modelo Busqueda a lo ancho
    def busquedaBFS(Grafos,lista): #grafos contiene los aristas y nodos del grafo, lista contiene el primer elemento de los nodos
            nodo_0 = lista.pop(0) #guarda el nodo inicial en nodo_0 y lo elimina de la lista
            #print("Nodo inicial " + nodo_0.getName())
            nodo_0.addRevisado() #etiqueda al nodo_0 como revisado
            nivel = nodo_0.getNivel() #marca el nivel del nodo
            conexiones = nodo_0.getConections() #obtiene las conecciones del nodo cero
            #print(conexiones)
            no_revisados = list()
            for nodo in conexiones:
                if not (nodo.getRevisado()):
                    lista.append(nodo)
                    no_revisados.append(nodo)
                    nodo.addRevisado()
                    nodo.addNivel(nivel+1)
                    nodo.addAncestro(nodo_0)
                #print("Conectado con: ")
                #for nodo in no_revisados:
                    #print(nodo.getName())  
                if not lista:
                    aristas = Grafo.reconstruir_arbol(Grafos[1])
                    return aristas 
                else:
                    return Grafo.busquedaBFS(Grafos,lista)

    # Modelo Busqueda en profundidad recursivo
    def busquedaDFSr(Grafos,lista):
            nodo_0 = lista.pop(0)
            nodo_0.addRevisado()
            conexiones = nodo_0.getConections()
            for nodo in conexiones:
                if not (nodo.getRevisado()):
                    lista.insert(0,nodo)
                    nodo.addAncestro(nodo_0)
            if not lista: # si la lista está vacía
                aristas = Grafo.reconstruir_arbol(Grafos[1])
                return aristas 
            else:
                return Grafo.busquedaDFSr(Grafos,lista)    


    # Modelo Busqueda en profundidad interativo
    def busquedaDFSi(Grafos,nodo_0):
            lista = [nodo_0]
            while lista: #mientras la lista no esté vacía
                nodo_0 = lista.pop(0)
                nodo_0.addRevisado()
                conexiones = nodo_0.getConections()
                for nodo in conexiones:
                    if not(nodo.getRevisado()):
                        lista.insert(0,nodo)
                        nodo.addAncestro(nodo_0)
            aristas = Grafo.reconstruir_arbol(Grafos[1])
            return aristas
    
    
    #Función para encontrar todas las aristas que inician en nodo a (sólo para grafos dirigidos)
    def buscar_conexiones(aristas,nodo,conectado):
        lista = list()
        if conectado == 1:
            for arista in aristas:
                ancestro = arista.getAncestro()
                if ancestro == nodo:
                    lista.append(arista)
        else:
            for arista in aristas:
                ancestro = arista.getAncestro()
                descendiente = arista.getDescendiente() 
                if ancestro == nodo or descendiente == nodo:
                    lista.append(arista)
        return lista
    
    def ordenar_prioridades(lista):
        for numPasada in range(len(lista)-1,0,-1):
            for i in range(numPasada):
                if lista[i].getPeso() > lista[i+1].getPeso():
                    temp = lista[i]
                    lista[i] = lista[i+1]
                    lista[i+1] = temp
        return lista

    
    
    def ordenar_prioridades_2(lista):
        for numPasada in range(len(lista)-1,0,-1):
            for i in range(numPasada):
                if lista[i].getPeso() < lista[i+1].getPeso():
                    temp = lista[i]
                    lista[i] = lista[i+1]
                    lista[i+1] = temp
        return lista    
    
    
    def ordenar_aristas(lista,m):
        if m == 0: #de menor a mayor
            for numPasada in range(len(lista)-1,0,-1):
                for i in range(numPasada):
                    if lista[i].getPeso() > lista[i+1].getPeso():
                        temp = lista[i]
                        lista[i] = lista[i+1]
                        lista[i+1] = temp
        else:
            for numPasada in range(len(lista)-1,0,-1):
                for i in range(numPasada):
                    if lista[i].getPeso() < lista[i+1].getPeso():
                        temp = lista[i]
                        lista[i] = lista[i+1]
                        lista[i+1] = temp
        return lista    
    

    def rastrear_camino(nodos,nodo_i,nodo_f):
        for nodo in nodos:
            try:
                ancestro = nodo.getAncestro()
                #print("Nodo " + nodo.getName() + " conectado a " + ancestro.getName())
                #nodo = ancestro
            except:
                print('')
                #print(nodo.getName()+' es el nodo inicial')
    
        
        
    
    def Dijkstra(nodos,aristas,nodo_0,nodo_f):
        S = list()
        PQ = nodos
        i = 0
        nodo_0.addPeso(0)
        acumulado_general = nodo_0.getPeso()
        while len(PQ) > 0:
            nodo_0 = PQ[0]
            index = PQ.index(nodo_0)
            acumulado_general = nodo_0.getPeso()
            nodo = PQ.pop(index)
            S.append(nodo)
            conexiones = Grafo.buscar_conexiones(aristas,nodo_0)
            #print('Nodo_0 '+ nodo_0.getName())
            #print('Conexiones ' + str(len(conexiones)))
            for arista in conexiones:
                nodo_1 = arista.getDescendiente()
                #print('Nodo_1 '+ nodo_1.getName())
                peso = nodo_1.getPeso()
                acumulado = acumulado_general + arista.getPeso()
                #print('Acumulado '+str(acumulado))
                #print('Peso '+str(peso))
                if acumulado < peso:
                    #print(nodo_1.getName()+' '+nodo_0.getName())
                    nodo_1.addPeso(acumulado)
                    nodo_1.addAncestro(nodo)
            PQ = Grafo.ordenar_prioridades(PQ)
        camino_corto = Grafo.rastrear_camino(S,nodo_0,nodo_f)
        aristas = Grafo.reconstruir_arbol(S)
        #print('longitud '+str(len(S)))
        return aristas


    
    def encontrar_ciclo(arbol,nodo_i,nodo_f):
        nodo_inicial = copy.deepcopy(nodo_i)
        aristas_guardadas = list()
        arista_guardada = None
        j = 0
        encontrado = 0
        while j <= len(nodos):
            j += 1
            i = 0
            while i <= len(nodos):
                i += 1
                contador = 0
                for pos in range(len(arbol)):
                    contador += 1
                    arista = arbol[pos]
                    nodo_1 = arista.getAncestro()
                    nodo_2 = arista.getDescendiente()
                    #print("Revisando si nodo " + nodo_1.getName() + " es igual a " + nodo_i.getName())
                    #print("Iguales " + str(nodo_1.getName() == nodo_i.getName()))
                    #print("Visitado " + str(not(arista.getVisitado())))
                    if (nodo_1.getName() == nodo_i.getName())and not(arista.getVisitado()):
                        nodo_i = arista.getDescendiente()
                        #print("Econtrado nodo " + nodo_1.getName() + " nuevo nodo inicial " + nodo_i.getName())
                        arista.addVisitado(True)
                        #print("Visitada la arista que conecta " + arista.getAncestro().getName() + " con " + arista.getDescendiente().getName())
                        arista_guardada = arista
                        break
                    #print("Revisando si nodo " + nodo_2.getName() + " es igual a " + nodo_i.getName())
                    #print("Iguales " + str(nodo_2.getName() == nodo_i.getName()))
                    #print("Visitado " + str(not(arista.getVisitado())))
                    if (nodo_2.getName() == nodo_i.getName())and not(arista.getVisitado()):
                        nodo_i = arista.getAncestro()
                        #print("Econtrado nodo " + nodo_2.getName() + " nuevo nodo inicial " + nodo_i.getName())
                        arista.addVisitado(True)
                        #print("Visitada la arista que conecta " + arista.getAncestro().getName() + " con " + arista.getDescendiente().getName())
                        arista_guardada = arista
                        break
                    if contador == len(arbol) and encontrado == 0:
                        nodo_i = nodo_inicial
                        #print("No se encontró, reiniciando nodo inicial a " + nodo_i.getName())
                        for n_arista in arbol:
                            aristas_guardadas.append(arista_guardada)
                            if n_arista not in aristas_guardadas:
                                n_arista.addVisitado(False)
                    if nodo_i == nodo_f:
                        encontrado = 1
                        break
                if encontrado == 1:
                    break
            if encontrado == 1:
                break
        return encontrado

  
                
   
    
    def Kruskal_D(aristas,nodos):
        lista = Grafo.ordenar_prioridades(aristas)
        arbol = list()
        for arista in lista:
            nodo_i = arista.getAncestro()
            nodo_f = arista.getDescendiente()
            encontrado = Grafo.encontrar_ciclo(arbol,nodo_i,nodo_f)
            if encontrado == 0:
                arbol.append(arista)
        return arbol
    
    
    #Enfoque de aristas
    #Se van eliminando las aristas más pesadas
    #Enfoque de aristas
    #Se van eliminando las aristas más pesadas
    def Kruskal_I(aristas,nodos):     
        lista = Grafo.ordenar_prioridades_2(aristas) # de mayor a menor
        lista_aux = lista
        arbol = list()       
        while len(arbol) < len(nodos)-1:
            arista = lista_aux.pop(0)
            nodo_i = arista.getAncestro()
            nodo_f = arista.getDescendiente()
            encontrado = Grafo.encontrar_ciclo(lista_aux,nodo_i,nodo_f)
            if encontrado == 0:
                arbol.append(arista)
                lista_aux.append(arista)       
        return arbol
    
    
    
    def prim(aristas,nodos,nodo_0):
        nodo1 = nodo_0
        azules = list()
        azules.append(nodo1)
        amarillos = list()
        revisados = list()
        arbol = list()
        i = 0

        while len(arbol) < len(nodos)-1:
            i += 1
            #print("Ciclo " + str(i))
            for azul in azules:
                #print("Buscando conexiones con nodo " + azul.getName())
                conexiones = Grafo.buscar_conexiones(aristas,azul,0)
                for arista in conexiones:
                    if arista not in amarillos and arista not in revisados:
                        if not arista.getAncestro() in azules or not arista.getDescendiente() in azules:
                            #print("Agregando conexión de nodo" + arista.getAncestro().getName() + " con nodo " + arista.getDescendiente().getName())
                            amarillos.append(arista)
            lista_amarilla = Grafo.ordenar_aristas(amarillos,0)
            #print("Lista ordenada")
            #for arista in lista_amarilla:
                #print("Nodo " + arista.getAncestro().getName() + " conectado con " + arista.getDescendiente().getName())
            nueva_arista = lista_amarilla[0]
            arbol.append(nueva_arista)
            revisados.append(nueva_arista)
            ancestro = nueva_arista.getAncestro()
            descendiente = nueva_arista.getDescendiente()

            if not ancestro in azules:
                #print("Agregando a azules el nodo " + ancestro.getName())
                azules.append(ancestro)
            if not descendiente in azules:
                #print("Agregando a azules el nodo " + descendiente.getName())
                azules.append(descendiente)

            index = amarillos.index(nueva_arista)
            amarillos.pop(index)

            #print("Revisando si hay que quitar alguna arista de amarillos")
            for arista in amarillos:
                ancestro1 = arista.getAncestro()
                descendiente1 = arista.getDescendiente()
                #print("Conexión de " + arista.getAncestro().getName() + " con " + arista.getDescendiente().getName())
                if ancestro1 in azules and descendiente1 in azules:
                    #print("Quitando conexion de " + ancestro1.getName() + " con " +  descendiente1.getName())
                    index = amarillos.index(arista)
                    amarillos.pop(index)

            #print("Azules")
            #for azul in azules:
            #    print(azul.getName())
            #print("Amarillos")
            #for arista in amarillos:
            #    print("Conexión de nodo " + arista.getAncestro().getName() + " con nodo " + arista.getDescendiente().getName())
            #print("Arbol")
            #for arista in arbol:
             #   print("Conexión de nodo " + arista.getAncestro().getName() + " con nodo " + arista.getDescendiente().getName())
        return arbol
    
    def ordenar(aristas,nodos,c1,c2,c4):
        xmax = 0
        ymax  = 0
        coordenadasx = np.zeros((len(nodos), 20))
        coordenadasy = np.zeros((len(nodos), 20))
        i = 0
        for nodo in nodos:
            nodo.addCoordenadas(random.randint(0, 50),random.randint(0, 500))
            x,y = nodo.getCoordenadas()
            coordenadasx[i][0] = x
            coordenadasy[i][0] = y
            i += 1

        for arista in aristas:
            nodo1 = arista.getAncestro()
            nodo2 = arista.getDescendiente()
            x1,y1 = nodo1.getCoordenadas()
            x2,y2 = nodo2.getCoordenadas()
            d = math.sqrt((x2-x1)**2+(y2-y1)**2)+0.0001
            force = c1*math.log(d/c2)
            nodo1.addCoordenadas(c4*force + x1,c4*force + y1)
            if (c4*force + x1) > xmax:
                xmax = c4*force + x1
            if (c4*force + y1) > ymax:
                ymax = c4*force + y1
            nodo2.addCoordenadas(c4*force + x2,c4*force + y2)
            if (c4*force + x2) > xmax:
                xmax = c4*force + x2
            if (c4*force + y2) > ymax:
                ymax = c4*force + y1
        propx = xmax/500
        propy = ymax/500
        for nodo in nodos:
            x,y = nodo.getCoordenadas()
            x2 = x/propx
            y2 = y/propy
            nodo.addCoordenadas(x2,y2)
        i = 0
        for nodo in nodos:
            x,y = nodo.getCoordenadas()
            xlin = np.linspace(coordenadasx[i][0], x, num=20)
            ylin = np.linspace(coordenadasy[i][0], y, num=20)
            for j in range(20):
                coordenadasx[i][j] = xlin[j]
                coordenadasy[i][j] = ylin[j]
            i += 1
        return aristas,nodos,coordenadasx,coordenadasy
```
